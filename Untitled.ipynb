```python
import pandas as pd
df = pd.read_csv('tubedata.csv', header=None)
df.head(50)
```

```python
from collections import defaultdict
 
station_dict = defaultdict(list)
zone_dict = defaultdict(set)
```

```python
# get data row by row
for index, row in df.iterrows():
  
  start_station = row[0]
  end_station = row[1]
  line = row[2]
  act_cost = int(row[3])
  zone1 = row[4]
  zone2 = row[5]
    
  # station dictionary of child station tuples (child_name, cost from parent to the child)
  # {"Mile End": [("Stepney Green", 2), ("Wembley", 1)]}
  station_list = station_dict[start_station]
  station_list.append((end_station, act_cost, line))

  # the following two lines add the other direction of the tube "step"
  station_list = station_dict[end_station]
  station_list.append((start_station, act_cost, line))
  
  # we add the main zone
  zone_dict[start_station].add(zone1)
  # we add the secondary zone
  if zone2 != "0":
#     print(row[2], start_station, end_station)
    zone_dict[start_station].add(zone2)
    # if the secondary zone is not 0 it's the main zone for the ending station
    zone_dict[end_station].add(zone2)
  else:
    # otherwise the main zone for the ending station is the same as for the starting station
    zone_dict[end_station].add(zone1)
```

```python
def dfs(graph, start, end, visited=None, path=None, cost=0, expanded=0, reverse=False):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path = path + [start]
    expanded += 1

    if start == end:
        # If the destination is reached, return the path, cost, and number of nodes expanded
        return path, cost, expanded
    
    graph_order = reversed(graph[start]) if reverse else graph[start]
    
    for neighbor, neighbor_cost,_ in graph_order:
        if neighbor not in visited:
            new_cost = cost + neighbor_cost
            result = dfs(graph, neighbor, end, visited, path, new_cost, expanded, reverse)
            if result:
                return result

    path.pop()
    visited.remove(start)
    
# Example usage:
path, cost, expanded = dfs(station_dict, "Euston", "Victoria")
print("Path ", path)
print("Cost ", cost)
print("Number of expanded nodes ", expanded)

path, cost, expanded = dfs(station_dict, "Euston", "Victoria", reverse = True)
print("\n\nPath ", path)
print("Cost ", cost)
print("Number of expanded nodes ", expanded)
```

```python
def bfs(graph, start, end):
    queue = [(start, [start], 0)]  # Initialize a list with (station, path, cost)
    visited = set()

    while queue:
        current_station, path, cost = queue.pop(0)  # Pop the first element
        visited.add(current_station)

        if current_station == end:
            # If the destination is reached, return the path, cost, and number of nodes expanded
            return path, cost, len(visited)

        for neighbor, neighbor_cost, _ in graph[current_station]:
            if neighbor not in visited:
                new_path = path + [neighbor]
                new_cost = cost + neighbor_cost
                queue.append((neighbor, new_path, new_cost))

    # If no path is found, return None
    return None, None, len(visited)


path, cost, expanded = bfs(station_dict, "Euston", "Victoria")
print("Path ", path)
print("Cost ", cost)
print("Number of expanded nodes ", expanded)
```

```python
def ucs(graph, start, end):
    queue = [(start, [start], 0)]  # Initialize a list with (station, path, cost)
    visited = set()

    while queue:
        queue.sort(key=lambda x: x[2])
        current_station, path, cost = queue.pop(0)  # Pop the first element
        visited.add(current_station)

        if current_station == end:
            # If the destination is reached, return the path, cost, and number of nodes expanded
            return path, cost, len(visited)

        for neighbor, neighbor_cost, _ in graph[current_station]:
            if neighbor not in visited:
                new_path = path + [neighbor]
                new_cost = cost + neighbor_cost
                queue.append((neighbor, new_path, new_cost))

    # If no path is found, return None
    return None, None, len(visited)


path, cost, expanded = ucs(station_dict, "Euston", "Victoria")
print("Path ", path)
print("Cost ", cost)
print("Number of expanded nodes ", expanded)
```

```python
def extenced_ucs(graph, start, end):
    queue = [(start, [start], 0, '')]  # Initialize a list with (station, path, cost)
    visited = set()

    while queue:
        queue.sort(key=lambda x: x[2])
#         print(queue)
        current_station, path, cost, current_line = queue.pop(0)  # Pop the first element
        visited.add(current_station)

        if current_station == end:
            # If the destination is reached, return the path, cost, and number of nodes expanded
            return path, cost, len(visited)

        for neighbor, neighbor_cost, line in graph[current_station]:
            if neighbor not in visited:
                new_path = path + [neighbor]
                new_cost = cost + neighbor_cost
                      
                if line != current_line:
                    print("line change cost from {} to {}".format(current_line, line))
                    new_cost += 2 #line_change
                    
                queue.append((neighbor, new_path, new_cost, line))

    # If no path is found, return None
    return None, None, len(visited)


path, cost, expanded = extenced_ucs(station_dict, "Canada Water", "Stratford")
print("\n\nPath ", path)
print("Cost ", cost)
print("Number of expanded nodes ", expanded)
```

```python
import heapq

def calculate_heuristic(station, target):
    # Calculate the heuristic value for a station based on the minimum zone change to the target zone
    station_zones = zone_dict.get(station, set())
    target_zones = zone_dict.get(target, set())
    
    differences = []
    
    for z_1 in station_zones:
        for z_2 in target_zones:
            differences.append(abs(int(z_2) - int(z_1)))
            
    return min(differences)

def best_first_search_with_heuristic(graph, start, goal, zone_dict):
   
    frontier = []  # Priority queue for BFS
    heapq.heappush(frontier, (0, start))  # Initialize the queue with the start station
    explored = set()  # Set to keep track of explored stations
    
    total_path_time = 0
    num_nodes_expanded = 0

    while frontier:
        _, current_station = heapq.heappop(frontier)  # Get the station with the lowest heuristic value
        num_nodes_expanded += 1
            
        if current_station == goal:
            return explored, total_path_time, num_nodes_expanded # Return the set of explored stations

        explored.add(current_station)  # Add the current station to the explored set

        for neighbor, time, _ in graph[current_station]:
            if neighbor not in explored:
                total_path_time += time
                heuristic_value = calculate_heuristic(neighbor, goal)  # Calculate the heuristic value for the neighbor station
                heapq.heappush(frontier, (heuristic_value, neighbor))  # Add neighbor to the frontier with its heuristic value

    return None  # No path found


# Perform the Best-First Search with the heuristic

path,total_path_time, num_nodes_expanded  = best_first_search_with_heuristic(station_dict, "Canada Water","Stratford", zone_dict)

# Print the explored stations
print("Path:", path)
print("Total Path Time:", total_path_time)
print("Number of Nodes Expanded:", num_nodes_expanded)
```
